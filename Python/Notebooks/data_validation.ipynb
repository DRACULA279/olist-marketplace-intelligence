Importing engine

```python
import pandas as pd 
import sys 
import os 

sys.path.append(os.path.abspath("../scripts"))

from db_connection import engine
```

DB to python


```python
table_names = ['customers','geolocation','order_items','order_payments', 'order_reviews','orders','products','product_category_name_translation','sellers']

tables = {}

for table in table_names:

    query = f"SELECT * FROM {table}"

    try:

        tables[table] = pd.read_sql(query,engine)

        print(f"The table {table} has been loaded successfully.")

    except Exception as error:
        print(f"The table {table} failed to load.")
        print(f"Reason: {error}")
```

validation

```python
for table_name, df in tables.items():

    print("\n" + "=" * 70)
    print(f"{table_name.upper()}")
    print("=" * 70)

    print(f"\nShape: {df.shape}")

    print("\nData Types:")
    print(df.dtypes)

    # Missing values
    null_counts = df.isnull().sum()

    null_percentages = (
        null_counts / len(df)
    ) * 100

    null_summary = pd.DataFrame({
        "Null Count": null_counts,
        "Null Percentage": null_percentages.round(2)
    })

    print("\nMissing Values:")

    print(
        null_summary[
            null_summary["Null Count"] > 0
        ]
    )

    # Duplicate rows
    duplicate_count = df.duplicated().sum()

    duplicate_percentage = (
        duplicate_count / len(df)
    ) * 100

    print("\nDuplicate Rows:")
    print(duplicate_count)

    print("\nDuplicate Percentage:")
    print(f"{duplicate_percentage:.2f}%")
```
